```python
from langgraph.graph import START, END, StateGraph
from langchain_openai.chat_models import ChatOpenAI
from dotenv import load_dotenv
import os
from langgraphagenticai.nodes.validate_data import validate_node
from langgraphagenticai.nodes.build_features import build_features_node
from langgraphagenticai.nodes.yahoo_data_extraction_node import yahoo_extract_node
from langgraphagenticai.state.commodity_state import CommodityResearchState
```

```python
load_dotenv()
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
```

```python
# from your_state_module import CommodityResearchState
graph = StateGraph(CommodityResearchState)

graph.add_node("yahoo_extract", yahoo_extract_node)
graph.add_node("build_features", build_features_node)
graph.add_node("validate", validate_node)

graph.add_edge(START, "yahoo_extract")
graph.add_edge("yahoo_extract", "build_features")
graph.add_edge("build_features", "validate")
graph.add_edge("validate", END)

app = graph.compile()
```
